## Unzip Dataset

```python
import zipfile

# Path to the downloaded zip
zip_path = "google-maps-restaurant-reviews.zip"

with zipfile.ZipFile(zip_path, 'r') as zip_ref:
    # Check all files inside
    print(zip_ref.namelist())

    # Extract only reviews.csv
    zip_ref.extract("reviews.csv", "./")
```
